```python
%matplotlib notebook
import torch
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from model import visutils
from model.selectpoints import select_k_points
```

```python
pts = []
for i in range(3,5):
    fig = plt.figure()
    ax1 = fig.add_subplot(121)
    ax2 = fig.add_subplot(122)
    img1 = plt.imread(f"./model/dtu_data/object/{i}.png")
    img2 = plt.imread(f"./model/dtu_data/object/{i+1}.png")
    ax1.imshow(img1)
    ax2.imshow(img2)
    pts1 = select_k_points(ax1, 3)
    pts2 = select_k_points(ax2, 3)
    pts.append([[pts1.xs, pts1.ys],[pts2.xs, pts2.ys]])
    plt.show()
```

```python
import numpy as np
pts = np.array(pts)
print(pts.shape)
```
